```python
import cv2
import numpy as np
from visualize_cv2 import model, display_instances, class_names

from matplotlib import pyplot as plt
```

```python
image = cv2.imread('./images/2516944023_d00345997d_z.jpg',1)
image_rgb =  cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```
